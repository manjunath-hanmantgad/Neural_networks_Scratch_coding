We just need to perform transposition on its second argument, which is the weights matrix in our case, to turn the row vectors it currently consists of into column vectors.

#### to perform the matrix product, the shape’s value from the index 1 of the first matrix and the index 0 of the second matrix must match.

#### value from index 1 means 1st matrix's all columns (only the 1st elements of all columns) 
#### value from index 0 means 2nd matrix's all rows (only the 1st element of all rows)

```python
import numpy as np
import nnfs
nnfs.init()

n_inputs = 2
n_neurons = 4

weights = 0.01 * np.random.randn(n_inputs, n_neurons)
biases = np.zeros((1, n_neurons)) #(1,n --> 1 row and zeros = n)
# https://numpy.org/doc/stable/reference/generated/numpy.zeros.html

print(weights)
print(biases)
```

```python
class Layer_Dense:
    def __init__(self,n_inputs,n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        # weights = n_inputs for 1st dimension and number of neurons for 2nd dimension 
      # similar to weights and inputs with transpose and other.
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
            self.output = np.dot(inputs , self.weights) + self.biases
        
    
```

```python
# create dataset
X, y = spiral_data(samples=100, classes=3)

# Create Dense layer with 2 input features and 3 output values
dense1 = Layer_Dense(2,3)

# Perform a forward pass of our training data through this layer
dense1.forward(X)


print(dense1.output[:5])
```
